```python
import pandas as pd
import numpy as np
import os

os.path.abspath("")
```

```python
data = pd.read_csv("realdonaldtrump.csv", encoding='latin1',index_col = False, error_bad_lines=False)
data.head()
```

# Checking for null values

```python
data.isnull().sum()
```

```python
data.drop(["link"], axis =1,inplace=True)
```

```python
#filtering data from 2012 onwards
df = data[(data['date'] > '2012-01-01') & (data['date'] < '2020-12-12')]
```

# Adding Sentimental analysis

```python
pip install -U textblob
```

```python
#Cleanning up tweet column
df['tidy_tweets'] = df['content'].str.replace('[^a-zA-Z#]+',' ')
df.drop(["content"], axis =1,inplace=True)
df
```

```python
# after the tweet column has been cleaned we used the tidy_tweet to apply the sentimental analyis
import pandas as pd
from textblob import TextBlob

df['sentiment'] = df.tidy_tweets.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

### Transforming sentimental column

```python
def applyFunc(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    return 'Neutral'

df['Overall_sentimental'] = df['sentiment'].apply(applyFunc)
df
```

### Creating Day of the week and time colums

```python
df['date'] = pd.to_datetime(df['date'])

df['day_of_week'] = df['date'].dt.day_name()
```

```python
df['date_only'] = df['date'].dt.date
```

```python
import datetime
# extracting time from timestamp
df['Time'] = [datetime.datetime.time(d) for d in df['date']] 
```

```python
# after extracting time and day_of_week from timestamp I drop created_at
df.drop(["date"], axis =1,inplace=True)
df.head(10)
```

```python
df.to_csv('DonaldTrump_Clean.csv',index=False)
```

## Creating a summary table by date and mean sentimental

```python
new_df = df.filter(['sentiment','date_only'], axis=1)
```

```python
new_df['date_only'] = pd.to_datetime(new_df['date_only'])
new_df
```

```python
date_only = new_df.set_index('date_only')
date_only
```

```python
#Monthly summary table with mean sentimental column - freq = M = Monthly check documentation for more info
monthly_df = date_only.groupby(pd.Grouper(freq='M', level='date_only')).mean()
```

```python
#Monthly summary table with mean sentimental column - freq = M = Monthly check documentation for more info
monthly_df["Average_sentiment"] = date_only.groupby(pd.Grouper(freq='M', level='date_only')).mean()
```

```python
monthly_df["count_tweets_per_day"] = date_only.groupby(pd.Grouper(freq='M', level='date_only')).count()
```

```python
#reseting index to default - RUN TIHS ONLY ONCE as it creates new index and level_o every time 
monthly_df.reset_index(level=0, inplace=True)
monthly_df.drop(["sentiment"], axis =1,inplace=True)
```

```python
monthly_df
```

```python
monthly_df.to_csv('DonaldTrump_Monthly_summary.csv',index=False)
monthly_df
```

### Daily summary table

```python
#Dainly summary table with mean sentimental column
daily_mean = date_only.groupby(pd.Grouper(freq='1d', level='date_only')).mean()
daily_mean
```

```python
#Dainly summary table with count of tweets per dar column
daily_count = date_only.groupby(pd.Grouper(freq='1d', level='date_only')).count()
daily_count
```

```python
daily_summary = pd.merge(daily_mean,daily_count, left_index=True, right_index=True)
daily_summary
```

```python
# Resetting index
daily_summary.reset_index(level=0, inplace=True)
daily_df = daily_summary.rename(columns={"sentiment_x": "mean_sentiment", "sentiment_y": "countOf_tweets_day"})
```

```python
daily_df.drop(["index"], axis =1,inplace=True)
```

```python
def applyFunc_2(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    return 'Neutral'

daily_df['tweet_sentiment'] = daily_df['mean_sentiment'].apply(applyFunc_2)
daily_df.to_csv('DonaldTrump_Daily_summary.csv',index=False)
daily_df
```
