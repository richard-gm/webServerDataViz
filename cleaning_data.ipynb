# Adding Sentimental analysis

```python
pip install -U textblob
```

```python
#Cleanning up tweet column
df['tidy_tweet'] = df['tweet'].str.replace('[^a-zA-Z#]+',' ')
df.drop(["tweet"], axis =1,inplace=True)
df
```

```python
# after the tweet column has been cleaned we used the tidy_tweet to apply the sentimental analyis
import pandas as pd
from textblob import TextBlob

df['sentiment'] = df.tidy_tweet.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
df
```

### Transforming sentimental column

```python
df.drop(["tidy_tweet"], axis =1,inplace=True)
df
```

```python
df['sentiment'].round(2)
df
```

### Creating Day of the week and time colums

```python
df['created_at'] = pd.to_datetime(df['created_at'])

df['day_of_week'] = df['created_at'].dt.day_name()
df
```

```python
import datetime
# extracting time from timestamp
df['Time'] = [datetime.datetime.time(d) for d in df['created_at']] 
df
```

```python
# after extracting time and day_of_week from timestamp I drop created_at
df.drop(["created_at"], axis =1,inplace=True)
df.head(10)
```
