```python
data = pd.read_csv("realdonaldtrump.csv", encoding='latin1',index_col = False, error_bad_lines=False)
data.head()
```

# Checking for null values

```python
data.isnull().sum()
```

```python
data.drop(["link"], axis =1,inplace=True)
```

# Adding Sentimental analysis

```python
pip install -U textblob
```

```python
#Cleanning up tweet column
data['tidy_tweets'] = data['content'].str.replace('[^a-zA-Z#]+',' ')
data.drop(["content"], axis =1,inplace=True)
data
```

```python
# after the tweet column has been cleaned we used the tidy_tweet to apply the sentimental analyis
import pandas as pd
from textblob import TextBlob

data['sentiment'] = data.tidy_tweets.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
data
```

### Creating Day of the week and time colums

```python
data['date'] = pd.to_datetime(data['date'])

data['day_of_week'] = data['date'].dt.day_name()
data
```

```python
data['date_only'] = data['date'].dt.date
```

```python
import datetime
# extracting time from timestamp
data['Time'] = [datetime.datetime.time(d) for d in data['date']] 
```

```python
# after extracting time and day_of_week from timestamp I drop created_at
data.drop(["date"], axis =1,inplace=True)
data.head(10)
```

### Transforming sentimental column

```python
def applyFunc(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    elif s == 0.000000:
        return 'Neutral'
    return ''

data['Overall_sentimental'] = data['sentiment'].apply(applyFunc)
data
```

```python
data.drop(["Overall_Sentimental"], axis =1,inplace=True)
```

```python
data.to_csv('DonaldTrump_Clean_tweets.csv',index=False)
```
