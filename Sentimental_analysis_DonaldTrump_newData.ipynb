```python
import pandas as pd
import numpy as np
import os

os.path.abspath("")
```

```python
data = pd.read_csv("DonalTrump_updated_data.csv", encoding='latin1',index_col = False, error_bad_lines=False)
data.head()
```

# Checking for null values

```python
data.isnull().sum()
```

```python
data.drop(["is_retweet","id_str"], axis =1,inplace=True)
```

```python
data['created_at'] = pd.to_datetime(data['created_at'])
data
```

# Adding Sentimental analysis

```python
pip install -U textblob
```

```python
#Cleanning up tweet column
data['tidy_tweets'] = data['text'].str.replace('[^a-zA-Z#]+',' ')
data.drop(["text"], axis =1,inplace=True)
data
```

```python
# after the tweet column has been cleaned we used the tidy_tweet to apply the sentimental analyis
import pandas as pd
from textblob import TextBlob

data['sentiment'] = data.tidy_tweets.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

### Transforming sentimental column

```python
def applyFunc(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    return 'Neutral'

data['Overall_sentimental'] = data['sentiment'].apply(applyFunc)
data
```

### Creating Day of the week and time colums

```python
data['date'] = pd.to_datetime(data['created_at'])

data['day_of_week'] = data['created_at'].dt.day_name()
```

```python
data['date_only'] = data['created_at'].dt.date
```

```python
import datetime
# extracting time from timestamp
data['Time'] = [datetime.datetime.time(d) for d in data['created_at']] 
```

```python
# after extracting time and day_of_week from timestamp I drop created_at
data.drop(["created_at","date"], axis =1,inplace=True)
data.head(10)
```

```python
data
```

```python
data.to_csv('DonaldTrump_upToDate.csv',index=False)
```

## Creating a summary table by date and mean sentimental

```python
new_df = data.filter(['sentiment','date_only'], axis=1)
```

```python
new_df['date_only'] = pd.to_datetime(new_df['date_only'])
new_df
```

```python
date_only = new_df.set_index('date_only')
date_only
```

```python
#Monthly summary table with mean sentimental column - freq = M = Monthly check documentation for more info
monthly_df = date_only.groupby(pd.Grouper(freq='M', level='date_only')).mean()
```

```python
#Monthly summary table with mean sentimental column - freq = M = Monthly check documentation for more info
monthly_df["Average_sentiment"] = date_only.groupby(pd.Grouper(freq='M', level='date_only')).mean()
```

```python
monthly_df["count_tweets_per_day"] = date_only.groupby(pd.Grouper(freq='M', level='date_only')).count()
```

```python
#reseting index to default - RUN TIHS ONLY ONCE as it creates new index and level_o every time 
monthly_df.reset_index(level=0, inplace=True)
monthly_df.drop(["sentiment"], axis =1,inplace=True)
```

```python
monthly_df
```

```python
monthly_df.to_csv('DonaldTrump_Monthly_summary.csv',index=False)
monthly_df
```

### Daily summary table

```python
#Dainly summary table with mean sentimental column
daily_mean = date_only.groupby(pd.Grouper(freq='1d', level='date_only')).mean()
daily_mean
```

```python
#Dainly summary table with count of tweets per dar column
daily_count = date_only.groupby(pd.Grouper(freq='1d', level='date_only')).count()
daily_count
```

```python
daily_summary = pd.merge(daily_mean,daily_count, left_index=True, right_index=True)
daily_summary
```

```python
# Resetting index
daily_summary.reset_index(level=0, inplace=True)
daily_df = daily_summary.rename(columns={"sentiment_x": "mean_sentiment", "sentiment_y": "countOf_tweets_day"})
```

```python
def applyFunc_2(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    return 'Neutral'

daily_df['tweet_sentiment'] = daily_df['mean_sentiment'].apply(applyFunc_2)
daily_df
```

### Adding fav and Number of tweets

```python
retweetsColumn_df = data.filter(['retweet_count','date_only'], axis=1)
favoritesColumn_df = data.filter(['favorite_count','date_only'], axis=1)

retweetsColumn_df['date_only'] = pd.to_datetime(new_df['date_only'])
favoritesColumn_df['date_only'] = pd.to_datetime(new_df['date_only'])
```

```python
retweetsByDate = retweetsColumn_df.set_index('date_only')
favoritesByDate = favoritesColumn_df.set_index('date_only')
favoritesByDate
```

```python
retweetsByDate
```

```python
#Dainly summary table with count of tweets per dar column
retweetDaily_count = retweetsByDate.groupby(pd.Grouper(freq='1d', level='date_only')).sum()
retweetDaily_count
```

```python
#Dainly summary table with count of tweets per dar column
favoritesDaily_count = favoritesByDate.groupby(pd.Grouper(freq='1d', level='date_only')).sum()
favoritesDaily_count
```

```python
ret_and_fav_summary = pd.merge(retweetDaily_count,favoritesDaily_count, left_index=True, right_index=True)
ret_and_fav_summary
```

```python
ret_and_fav_summary.reset_index(level=0, inplace=True)
ret_and_fav_summary
```

```python
final_summary = pd.merge(daily_df,ret_and_fav_summary, left_index=True, right_index=True)
final_summary
```

```python
final_summary.to_csv('DonaldTrump_Daily_Updated.csv',index=False)
```
