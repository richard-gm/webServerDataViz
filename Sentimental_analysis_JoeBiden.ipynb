```python
import pandas as pd
import numpy as np
import os

os.path.abspath("")
```

```python
data = pd.read_csv("JoeBidenTweets.csv", encoding='latin1',index_col = False, error_bad_lines=False)
data.head()
```

# Checking for null values

```python
data.isnull().sum()
```

```python
data.drop(["url"], axis =1,inplace=True)
```

# Adding Sentimental analysis

```python
pip install -U textblob
```

```python
#Cleanning up tweet column
data['tidy_tweets'] = data['tweet'].str.replace('[^a-zA-Z#]+',' ')
data.drop(["tweet"], axis =1,inplace=True)
data
```

```python
# after the tweet column has been cleaned we used the tidy_tweet to apply the sentimental analyis
import pandas as pd
from textblob import TextBlob

data['sentiment'] = data.tidy_tweets.apply(lambda x: TextBlob(str(x)).sentiment.polarity)
data
```

### Transforming sentimental column

```python
def applyFunc(s):
    if s >= 0.1:
        return 'Positive'
    elif s <= -0.1:
        return 'Negative'
    return 'Neutral'

data['Overall_sentimental'] = data['sentiment'].apply(applyFunc)
data
```

### Creating Day of the week and time colums

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])

data['day_of_week'] = data['timestamp'].dt.day_name()
data
```

```python
data['date_only'] = data['timestamp'].dt.date
```

```python
import datetime
# extracting time from timestamp
data['Time'] = [datetime.datetime.time(d) for d in data['timestamp']] 
```

```python
# after extracting time and day_of_week from timestamp I drop created_at
data.drop(["timestamp"], axis =1,inplace=True)
data.head(10)
```

```python
data.dtypes
```

## Creating a summary table by date and mean sentimental

```python
new_df = data.filter(['sentiment','date_only'], axis=1)
new_df
```

```python
new_df['date_only'] = pd.to_datetime(new_df['date_only'])
date_only = new_df.set_index('date_only')
```

```python
#Monthly summary table with mean sentimental column
monthly_df = date_only.groupby(pd.Grouper(freq='M', level='date_only')).mean()
```

```python
#reseting index to default - RUN TIHS ONLY ONCE as it creates new index and level_o every time 
monthly_df.reset_index(level=0, inplace=True)
```

```python
#filtering data from 2012 onwards
monthly_dataframe = monthly_df[(monthly_df['date_only'] > '2012-01-01') & (monthly_df['date_only'] < '2020-12-12')]
```

```python
monthly_dataframe.to_csv('JoeBiden_Monthly_summary.csv',index=False)
monthly_dataframe
```

### Daily summary table

```python
#Dainly summary table with mean sentimental column
daily_df = date_only.groupby(pd.Grouper(freq='1d', level='date_only')).mean()
```

```python
#filtering data from 2012 onwards - move this to the beggining
daily_df.reset_index(level=0, inplace=True)
daily_dataFrame = daily_df[(daily_df['date_only'] > '2012-01-01') & (daily_df['date_only'] < '2020-12-12')]
```

```python
daily_dataFrame.to_csv('JoeBiden_Daily_summary.csv',index=False)
daily_dataFrame
```
